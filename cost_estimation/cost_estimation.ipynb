```python
import sliding_window
import os
import json
```

```python
ORIGINAL_PATH = "dataset/shortest_unlabeled"
file_paths = os.listdir(ORIGINAL_PATH)
contracts = [os.path.basename(file_path)[:-4] for file_path in file_paths]
WINDOW_SIZE = 6000

window_count = 0
for contract in contracts:
    window_indexes = sliding_window.get_sliding_window_indexes(contract, WINDOW_SIZE, 0, ORIGINAL_PATH)
    window_count+= len(window_indexes)


total_token = window_count * (25000*2.5 + 8000*10)
total_token/1000000
```

```python
ORIGINAL_PATH = "dataset/unified_LLM"
split_paths = "dataset/LLM_split_paths_merge.json"
with open(split_paths, 'r') as f:
    config = json.load(f)
test_files = config.get("test", [])
test_file_names = [os.path.basename(file_path)[:-5] for file_path in test_files]
window_count = 0
for contract in test_file_names:
    window_indexes = sliding_window.get_sliding_window_indexes(contract, WINDOW_SIZE, 0, ORIGINAL_PATH)
    window_count+= len(window_indexes)


total_token = window_count * (25000*2.5 + 8000*10)
total_token/1000000
```
